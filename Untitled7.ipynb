4. Data Preparation


4.1 Drop id Column & Convert Age

```python




df = df.drop(columns=['id', 'age'])
df = df.rename(columns={'age_years': 'age'})
print(f'Shape after dropping id & converting age: {df.shape}')
df.head()



```


4.2 Outlier Removal — Blood Pressure
Outlier removal on blood pressure — remove rows with ap_hi / ap_lo outside reasonable physiological ranges (systolic 80–250 mmHg, diastolic 40–180 mmHg, and ap_hi >= ap_lo). These extreme outliers are almost certainly data entry errors.

```python


before = df.shape[0]
df = df[(df['ap_hi'].between(80, 250)) & (df['ap_lo'].between(40, 180))]
df = df[df['ap_hi'] >= df['ap_lo']]
after = df.shape[0]
print(f'Rows before: {before} | after blood pressure filter: {after} | dropped: {before - after}')

```


4.3 Outlier Removal — Height & Weight (1%–99% percentile

```python
before = df.shape[0]
for col in ['height', 'weight']:
    low, high = df[col].quantile([0.01, 0.99])
    df = df[df[col].between(low, high)]
after = df.shape[0]
print(f'Rows before: {before} | after height/weight filter: {after} | dropped: {before - after}')
```


4.4 Drop Duplicates

```python


before = df.shape[0]
df = df.drop_duplicates().reset_index(drop=True)
after = df.shape[0]
print(f'Rows before: {before} | after dropping duplicates: {after} | dropped: {before - after}')
print(f'\nFinal dataset shape: {df.shape}')

```


4.5 BMI Feature Engineering

Body Mass Index (BMI) = $\frac{weight}{(height / 100)^2}$ is a clinically established obesity indicator and a known CVD risk factor. Adding it provides a more direct representation of body composition than using weight and height separately.


```python


df['bmi'] = (df['weight'] / (df['height'] / 100) ** 2).round(2)

print(f"BMI feature added. Stats:")
print(df['bmi'].describe().round(2))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df['bmi'], bins=40, kde=True, ax=axes[0], color='#3498db')
axes[0].set_title('BMI Distribution')
axes[0].set_xlabel('BMI')

sns.boxplot(x='cardio', y='bmi', data=df, ax=axes[1], palette=['#2ecc71', '#e74c3c'])
axes[1].set_title('BMI vs Cardiovascular Disease')
axes[1].set_xlabel('cardio (0 = Healthy, 1 = CVD)')
axes[1].set_ylabel('BMI')
plt.tight_layout()
plt.show()

```


4.6 Split Features and Target

```python
X = df.drop(columns=['cardio'])
y = df['cardio']
print(f'X shape: {X.shape} | y shape: {y.shape}')
```


4.7 Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
print(f'Train size: {X_train.shape[0]} | Test size: {X_test.shape[0]}')
```


4.8 Standardization

```python

numeric_cols = ['age', 'height', 'weight', 'ap_hi', 'ap_lo', 'bmi']
scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])

X_train_scaled.head()
```



5. Modeling




5.1 Baseline — Logistic Regression

```python
logreg = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
logreg.fit(X_train_scaled, y_train)
y_pred_lr = logreg.predict(X_test_scaled)


```


5.2 Baseline — Random Forest Classifie

```python
rf = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=-1)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
```

5.3 Hyperparameter Tuning — Random Forest (GridSearchCV)
The grid is chosen to be small & specific so that the runtime remains affordable on a large dataset.

```python
param_grid = {
    'n_estimators': [200],
    'max_depth': [10, 15, 20],
    'min_samples_leaf': [2, 5],
}

grid = GridSearchCV(
    estimator=RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1),
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=1,
)
grid.fit(X_train_scaled, y_train)

print(f'Best params: {grid.best_params_}')
print(f'Best CV F1-score: {grid.best_score_:.4f}')

rf_tuned = grid.best_estimator_
y_pred_rf_tuned = rf_tuned.predict(X_test_scaled)
```

```python
def evaluate_model(name, y_true, y_pred, y_proba=None):
    return {
        'Model': name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'ROC-AUC': roc_auc_score(y_true, y_proba) if y_proba is not None else np.nan,
    }

results = pd.DataFrame([
    evaluate_model('Logistic Regression', y_test, y_pred_lr, logreg.predict_proba(X_test_scaled)[:, 1]),
    evaluate_model('Random Forest (Baseline)', y_test, y_pred_rf, rf.predict_proba(X_test_scaled)[:, 1]),
    evaluate_model('Random Forest (Tuned)', y_test, y_pred_rf_tuned, rf_tuned.predict_proba(X_test_scaled)[:, 1]),
])
results.set_index('Model').round(4)
```


6.1 Confusion Matrix


6.4 Feature Importance (Random Forest Tuned)

```python

from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

```

```python
print(X_train_scaled.shape)
print(y_train.shape)
```

```python
common_index = X_train_scaled.index.intersection(y_train.index)

X_train_scaled = X_train_scaled.loc[common_index]
y_train = y_train.loc[common_index]
```

```python
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train_scaled, y_train)
print(f'Avant SMOTE : {dict(zip(*np.unique(y_train, return_counts=True)))}')
print(f'Apres SMOTE : {dict(zip(*np.unique(y_train_sm, return_counts=True)))}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Distribution des classes — Avant vs Apres SMOTE', fontweight='bold')
pd.Series(y_train).value_counts().plot(
    kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'],
    edgecolor='white', title='Avant SMOTE'
)
pd.Series(y_train_sm).value_counts().plot(
    kind='bar', ax=axes[1], color=['#2ecc71', '#e74c3c'],
    edgecolor='white', title='Apres SMOTE'
)
for ax in axes:
    ax.set_xticklabels(['Healthy (0)', 'Cardio (1)'], rotation=0)
    ax.set_ylabel('Nombre de patients')
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# 7.1 DEFINITION DU MODELE ANN
# ============================================================
print('\nEntrainement du modele ANN...')

tf.random.set_seed(RANDOM_STATE)

model_ann = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_sm.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dropout(0.2),

    Dense(1, activation='sigmoid')  # Sortie binaire
])

model_ann.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['AUC']
)

early_stop = EarlyStopping(
    monitor='val_auc',
    patience=10,
    restore_best_weights=True,
    mode='max' )              # ← obligatoire car on veut maximiser

# Entrainement sur donnees SMOTE
history = model_ann.fit(
    X_train_sm, y_train_sm,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=0
)
print(f'ANN entraine (arret a l\'epoque {len(history.history["loss"])})')
```

courbe dapprentissage

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Loss
axes[0].plot(history.history['loss'],     label='Train Loss',      linewidth=2, color='#3498db')
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2, color='#e74c3c', linestyle='--')
axes[0].set_xlabel('Epoque', fontsize=11)
axes[0].set_ylabel('Loss', fontsize=11)
axes[0].set_title("Courbe d'entrainement — Loss", fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# AUC — Keras peut nommer 'auc' ou 'AUC' selon la version
auc_key     = 'auc'     if 'auc'     in history.history else 'AUC'
val_auc_key = 'val_auc' if 'val_auc' in history.history else 'val_AUC'

axes[1].plot(history.history[auc_key],     label='Train AUC',      linewidth=2, color='#3498db')
axes[1].plot(history.history[val_auc_key], label='Validation AUC', linewidth=2, color='#e74c3c', linestyle='--')
axes[1].set_xlabel('Epoque', fontsize=11)
axes[1].set_ylabel('AUC', fontsize=11)
axes[1].set_title("Courbe d'entrainement — AUC", fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

prediction et seuil optimal

```python
# Predictions sur le test set
y_proba_ann = model_ann.predict(X_test_scaled, verbose=0).flatten()

# Seuil optimal (Youden Index : maximise tpr - fpr)
fpr_ann, tpr_ann, thresholds_ann = roc_curve(y_test, y_proba_ann)
best_threshold = thresholds_ann[np.argmax(tpr_ann - fpr_ann)]
y_pred_ann = (y_proba_ann > best_threshold).astype(int)

recall_ann = recall_score(y_test, y_pred_ann)
auc_ann    = roc_auc_score(y_test, y_proba_ann)

print(f'Seuil optimal  : {best_threshold:.4f}')
print(f'Recall         : {recall_ann:.4f}')
print(f'AUC            : {auc_ann:.4f}')
```

matrice de confusion ANN

```python
cm_ann = confusion_matrix(y_test, y_pred_ann)
cm_pct = cm_ann.astype(float) / cm_ann.sum(axis=1, keepdims=True) * 100
annot  = np.array([[f'{v}\n({p:.1f}%)' for v, p in zip(rv, rp)]
                    for rv, rp in zip(cm_ann, cm_pct)])

fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(
    cm_ann, annot=annot, fmt='', cmap='Blues', ax=ax,
    xticklabels=['Healthy', 'Cardio'],
    yticklabels=['Healthy', 'Cardio'],
    linewidths=0.5, linecolor='white', cbar=False
)
ax.set_title(f'Confusion Matrix — ANN\nRecall={recall_ann:.2f} | AUC={auc_ann:.2f}', fontweight='bold')
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
plt.tight_layout()
plt.show()

print('\n' + '='*60)
print('CLASSIFICATION REPORT — ANN')
print('='*60)
print(classification_report(y_test, y_pred_ann, target_names=['Healthy', 'Cardiovascular Disease']))
```
